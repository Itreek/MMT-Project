## South Familiarity

```python
# Familiarity of KOR for South participants

KOR_familiarity_mean = []

for index, row in south_hs.iterrows():
    # get the 'Language' columns that have the value 'KOR'
    kor_cols = [
        col for col in row.index if 'Language' in col and row[col] == 'KOR']

    # extract the 'Familiarity' values corresponding to the 'Language' columns
    kor_familiarities = [
        row[col.replace('Language', 'Familiarity')] for col in kor_cols]

    # get the mode of the 'Familiarity' values
    # familiarity_modes.append(np.argmax(np.bincount(kor_familiarities)))
    KOR_familiarity_mean.append(np.mean(kor_familiarities))

print(KOR_familiarity_mean)
```
