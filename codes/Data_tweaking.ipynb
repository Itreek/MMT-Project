# Co - relations

```python
import seaborn as sns
import numpy as np
from scipy import stats

# Example data
# familiarity = [3, 2, 4, 5, 1]
# preference = [7, 5, 8, 9, 3]
data = pd.read_csv('North scatter plots.csv')
familiarity = data['Familiarity']
preference = data['Preference']

# Compute the correlation coefficient and p-value
corr, p_val = np.corrcoef(familiarity, preference)[
    0, 1], stats.pearsonr(familiarity, preference)[1]

# Create scatter plot with linear model fitting
sns.lmplot(x='familiarity', y='preference', data=pd.DataFrame(
    {'familiarity': familiarity, 'preference': preference}))

# Print the results
print("Correlation coefficient:", corr)
print("p-value:", p_val)
```

```python
block1_south_hs = block1.loc[df['Home State'] == 'South']
block1_south_hs
```

#### Dreamy

```python
for_cor = []

for index, row in block1_south_hs.iterrows():
    # get the 'Language' columns that have the value 'KOR'
    rowwise_cor = []
    dreamy_cols = [
        col for col in row.index if '[Dreamy]' in col]
    
    for i in dreamy_cols:
        rowwise_cor.append(row[i])
    
    for_cor.append(rowwise_cor)
    
```

```python
for_cor
for_cor = pd.DataFrame(for_cor)
for_cor = for_cor.transpose()
for_cor.columns = ['rater1', 'rater2', 'rater3', 'rater4',
                     'rater5', 'rater6', 'rater7', 'rater8', 'rater9']

for_cor
```

```python
for_cor['index'] = for_cor.index
for_cor = pd.melt(for_cor, id_vars=['index'], value_vars=list(for_cor)[:-1])

print(for_cor)
```

```python
results = pg.intraclass_corr(for_cor, 'index', 'variable', 'value')
results = results.set_index('Description')
icc = results.loc['Average random raters', 'ICC']

print(icc.round(3))
```
